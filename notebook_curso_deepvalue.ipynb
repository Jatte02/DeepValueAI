## 1.1 Variables y Tipos de Datos

## 1.3 Condicionales (if / elif / else)

Los condicionales permiten que tu código tome **decisiones**. En trading, esto es fundamental:
- ¿El precio está por debajo de la media móvil? → Comprar
- ¿La pérdida supera el 12%? → Vender (stop loss)

La estructura es:
```python
if condición:
    # se ejecuta si la condición es True
elif otra_condición:
    # se ejecuta si la primera fue False y esta es True
else:
    # se ejecuta si ninguna condición fue True
```

**Importante:** Python usa **indentación** (4 espacios) para definir bloques de código, no llaves `{}`.

````python
# Ejemplo: lógica de stop loss y take profit
entry_price = 100.0
current_price = 85.0
stop_loss_pct = -0.12       # -12%
take_profit_pct = 0.15      # +15%

# Calcular el retorno actual
return_pct = (current_price - entry_price) / entry_price
print(f"Retorno actual: {return_pct:.2%}")

# Tomar decisión
if return_pct <= stop_loss_pct:
    action = "VENDER (Stop Loss)"
elif return_pct >= take_profit_pct:
    action = "VENDER (Take Profit)"
else:
    action = "MANTENER"

print(f"Decisión: {action}")

# Prueba cambiando current_price a 120.0 y ejecuta de nuevo
````

## 1.4 Loops (for y while)

Los **loops** repiten una acción múltiples veces. En finanzas los usamos para:
- Recorrer una lista de tickers y descargar datos de cada uno
- Evaluar señales de compra/venta día por día
- Iterar sobre todas las empresas del S&P 500

`for` recorre una secuencia. `while` repite mientras una condición sea True.

````python
# FOR LOOP — recorrer una lista de empresas
tickers = ["AAPL", "MSFT", "GOOGL", "NVDA"]
prices = [178.50, 420.00, 175.30, 950.00]

print("=== Portafolio ===")
for i, ticker in enumerate(tickers):
    # enumerate() te da el índice (i) y el valor (ticker) al mismo tiempo
    print(f"  {i+1}. {ticker}: ${prices[i]:,.2f}")

# FOR LOOP con range() — generar números
print("\n=== Primeros 5 días de trading ===")
for day in range(1, 6):    # range(1, 6) genera: 1, 2, 3, 4, 5
    print(f"  Día {day}")

# FOR LOOP con diccionario
metrics = {"sharpe_ratio": 1.94, "max_drawdown": -0.178, "win_rate": 0.698}

print("\n=== Métricas del modelo ===")
for name, value in metrics.items():
    print(f"  {name}: {value}")
````

## 1.5 Funciones

Una **función** es un bloque de código reutilizable con un nombre. En vez de copiar y pegar
el mismo cálculo 100 veces, lo defines una vez y lo llamas cuando lo necesites.

**¿Por qué importa?** En DeepValueAI, cada pieza lógica es una función:
- `download_ohlcv()` — descarga precios
- `compute_technical_features()` — calcula indicadores
- `predict_proba()` — genera predicciones

**Mejores prácticas:**
- Cada función debe hacer **una sola cosa**
- El nombre debe describir **qué hace** (verbo + sustantivo)
- Usa `return` para devolver resultados (no solo `print`)

````python
# Definir funciones reutilizables
def calculate_return(entry_price, current_price):
    """Calcula el retorno porcentual de una inversión."""
    return (current_price - entry_price) / entry_price

def classify_signal(probability, threshold=0.50):
    """Clasifica una probabilidad como BUY o HOLD según el umbral."""
    if probability >= threshold:
        return "BUY"
    return "HOLD"

def evaluate_position(entry_price, current_price, stop_loss=-0.12, take_profit=0.15):
    """Evalúa si una posición debe cerrarse."""
    ret = calculate_return(entry_price, current_price)  # reutilizamos la función anterior
    if ret <= stop_loss:
        return "SELL (Stop Loss)", ret
    if ret >= take_profit:
        return "SELL (Take Profit)", ret
    return "HOLD", ret

# Usar las funciones
print(f"Retorno AAPL: {calculate_return(150, 178):.2%}")
print(f"Señal (prob=0.75): {classify_signal(0.75)}")
print(f"Señal (prob=0.30): {classify_signal(0.30)}")

action, ret = evaluate_position(100, 85)
print(f"\nPosición: {action} (retorno: {ret:.2%})")
````

## 2.1 numpy — Operaciones Matemáticas Rápidas

`numpy` es la base de toda la ciencia de datos en Python. Permite trabajar con **arrays**
(vectores y matrices) de forma ultra-rápida — hasta 100x más rápido que loops de Python.

**¿Por qué importa?** Cuando calculas el RSI de 3,600 acciones con 5 años de datos,
necesitas procesar millones de números. numpy lo hace en milisegundos.

````python
import numpy as np

# Crear arrays
prices = np.array([150.0, 155.0, 148.0, 160.0, 162.0, 158.0, 165.0])
print(f"Precios: {prices}")
print(f"Tipo: {type(prices)}")

# Operaciones vectorizadas — se aplican a TODOS los elementos a la vez
returns = np.diff(prices) / prices[:-1]  # retornos diarios
print(f"\nRetornos diarios: {returns.round(4)}")

# Estadísticas en una línea
print(f"\nEstadísticas:")
print(f"  Media: {prices.mean():.2f}")
print(f"  Desviación estándar: {prices.std():.2f}")
print(f"  Mínimo: {prices.min():.2f}")
print(f"  Máximo: {prices.max():.2f}")
print(f"  Retorno promedio diario: {returns.mean():.4f} ({returns.mean()*100:.2f}%)")

# Comparación: ¿Qué días el precio superó 155?
above_155 = prices > 155
print(f"\n¿Precio > 155? {above_155}")
print(f"Días por encima: {above_155.sum()} de {len(prices)}")
````

## 3.1 ¿Qué es Machine Learning?

**Machine Learning (ML)** es enseñarle a una computadora a tomar decisiones
basándose en datos históricos, sin programar las reglas explícitamente.

En DeepValueAI, el ML funciona así:

```
ENTRADA (34 features)              SALIDA
────────────────────               ──────
RSI = 25                  ┌───┐
Williams %R = -85         │   │
PE Ratio = 12.5           │ M │──→  Probabilidad: 0.78 (78%)
VIX = 18.2         ──→    │ L │──→  Decisión: COMPRAR
Fed Rate = 5.25           │   │
Sentiment = 0.65          └───┘
...34 features total
```

**Clasificación vs Regresión:**
- **Clasificación** (lo que usamos): ¿Subirá o no? → Respuesta: Sí (1) o No (0)
- **Regresión**: ¿Cuánto subirá? → Respuesta: +12.5%

Usamos clasificación porque es más robusto — predecir la dirección es más fácil que predecir el número exacto.

## 3.2 Preparar los Datos

**Regla de oro del ML:** NUNCA entrenes con datos del futuro.

En finanzas esto es crítico. Si entrenas con datos de 2024 y luego evalúas con datos de 2023,
el modelo ya "vio el futuro" — esto se llama **look-ahead bias** y da resultados falsos.

**La solución:** Split temporal.
```
|────── Train (70%) ──────|── Val (15%) ──|── Test (15%) ──|
2020                      2023            2024             2025
```

````python
from core.config import FEATURE_COLUMNS

# Preparar los datos exactamente como lo hace DeepValueAI
df_sorted = df.sort_values("date").reset_index(drop=True)

# Reemplazar infinitos con NaN (algunos indicadores pueden dar inf)
df_sorted[FEATURE_COLUMNS] = df_sorted[FEATURE_COLUMNS].replace([np.inf, -np.inf], np.nan)

# Split temporal: 70% train, 15% validation, 15% test
n = len(df_sorted)
train_end = int(n * 0.70)
val_end = int(n * 0.85)

train = df_sorted.iloc[:train_end]
val = df_sorted.iloc[train_end:val_end]
test = df_sorted.iloc[val_end:]

print(f"Train: {len(train):>10,} filas ({train['date'].min().date()} → {train['date'].max().date()})")
print(f"Val:   {len(val):>10,} filas ({val['date'].min().date()} → {val['date'].max().date()})")
print(f"Test:  {len(test):>10,} filas ({test['date'].min().date()} → {test['date'].max().date()})")

# Separar features (X) de la variable objetivo (y)
X_train = train[FEATURE_COLUMNS].values
y_train = train["label"].values
X_val = val[FEATURE_COLUMNS].values
y_val = val["label"].values
X_test = test[FEATURE_COLUMNS].values
y_test = test["label"].values

print(f"\nFeatures: {len(FEATURE_COLUMNS)}")
print(f"Forma de X_train: {X_train.shape}  (filas × features)")
print(f"Balance en train: {y_train.mean():.2%} positivo")
````

## 3.3 Entrenar un Modelo

Vamos a entrenar el mismo modelo que usa DeepValueAI: **HistGradientBoostingClassifier**.

**¿Por qué este modelo?**
- Maneja NaN nativamente (muchas features tienen datos faltantes)
- Es rápido de entrenar (millones de filas en minutos)
- Es una implementación de sklearn equivalente a LightGBM/XGBoost
- No requiere preprocesamiento (normalización, imputación)

**¿Cómo funciona?** Construye muchos "árboles de decisión" pequeños, donde cada
árbol nuevo corrige los errores del anterior. Es como un equipo donde cada miembro
aprende de los errores de los demás.

````python
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import (
    precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report,
)

# Entrenar el modelo (¡esto es todo! 3 líneas)
model = HistGradientBoostingClassifier(
    max_iter=200,            # número de árboles
    max_depth=5,             # profundidad máxima de cada árbol
    learning_rate=0.05,      # qué tan rápido aprende (menor = más cuidadoso)
    min_samples_leaf=100,    # mínimo de muestras por hoja (evita overfitting)
    random_state=42,         # semilla para reproducibilidad
)

print("Entrenando modelo...")
model.fit(X_train, y_train)
print("¡Modelo entrenado!")

# Predecir probabilidades en el set de validación
y_val_proba = model.predict_proba(X_val)[:, 1]  # probabilidad de clase 1 (sube)

print(f"\nPrimeras 10 probabilidades predichas:")
for i in range(10):
    real = "SUBE" if y_val[i] == 1 else "NO SUBE"
    print(f"  Prob: {y_val_proba[i]:.4f} | Real: {real}")
````

## 3.4 Métricas de Evaluación — ¿Es bueno el modelo?

El modelo da una **probabilidad** (0 a 1). Necesitamos un **threshold** (umbral) para decidir:
- Si probabilidad ≥ threshold → predecir "SUBE" (1)
- Si probabilidad < threshold → predecir "NO SUBE" (0)

**¿Qué métricas importan?**

| Métrica | Significado | En trading |
|---------|-------------|------------|
| **Precision** | De las que predije "SUBE", ¿cuántas realmente subieron? | Si es baja, compras basura |
| **Recall** | De todas las que realmente subieron, ¿cuántas detecté? | Si es baja, pierdes oportunidades |
| **F1** | Balance entre precision y recall | Métrica general |
| **AUC** | Capacidad del modelo de separar las clases | 0.5 = random, 1.0 = perfecto |

En DeepValueAI priorizamos **precision** sobre recall — preferimos perder oportunidades
a comprar acciones que no suben. **Calidad sobre cantidad.**

````python
# Evaluar con diferentes thresholds para entender el tradeoff
print("Threshold | Precision | Recall | F1     | Predicciones BUY")
print("-" * 65)

for threshold in [0.30, 0.40, 0.50, 0.60, 0.70, 0.80]:
    y_pred = (y_val_proba >= threshold).astype(int)
    n_buy = y_pred.sum()
    if n_buy == 0:
        print(f"  {threshold:.2f}    |    N/A    |  N/A   |  N/A   | {n_buy:,}")
        continue
    prec = precision_score(y_val, y_pred, zero_division=0)
    rec = recall_score(y_val, y_pred, zero_division=0)
    f1 = f1_score(y_val, y_pred, zero_division=0)
    print(f"  {threshold:.2f}    |  {prec:.4f}  | {rec:.4f} | {f1:.4f} | {n_buy:,}")

print("\n→ Observa: threshold alto = más precisión pero menos señales")
print("→ DeepValueAI optimiza el threshold para máxima precision con recall ≥ 25%")
````

## 4.2 Indicadores Técnicos

Los **indicadores técnicos** son fórmulas matemáticas que transforman datos de precio/volumen
en señales que ayudan a predecir movimientos futuros.

DeepValueAI usa 11 indicadores técnicos. Aquí los más importantes:

| Indicador | Qué mide | Señal de compra |
|-----------|----------|-----------------|
| **SMA 200** | Tendencia de largo plazo | Precio cerca o debajo de la SMA |
| **RSI** | ¿Sobrecomprada o sobrevendida? | RSI < 30 (sobrevendida) |
| **Williams %R** | Momentum | Cruza de -80 hacia arriba |
| **MACD** | Cambio de tendencia | Histograma cruza de negativo a positivo |

````python
# Calcular indicadores técnicos manualmente (para entender la lógica)

# SMA 200 — promedio de los últimos 200 días de precio de cierre
data["SMA_200"] = data["Close"].rolling(window=200).mean()

# RSI — Relative Strength Index (14 días)
delta = data["Close"].diff()
gain = delta.where(delta > 0, 0).rolling(14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(14).mean()
rs = gain / loss
data["RSI"] = 100 - (100 / (1 + rs))

# Distancia del precio a la SMA 200 (feature clave de DeepValueAI)
data["SMA_Distance"] = (data["Close"] - data["SMA_200"]) / data["SMA_200"]

# Visualizar
fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=True)

# Precio + SMA
axes[0].plot(data.index, data["Close"], label="Precio", linewidth=1)
axes[0].plot(data.index, data["SMA_200"], label="SMA 200", linewidth=1.5, color="orange")
axes[0].set_title("AAPL — Precio vs SMA 200")
axes[0].legend()

# RSI
axes[1].plot(data.index, data["RSI"], color="purple", linewidth=1)
axes[1].axhline(y=70, color="red", linestyle="--", alpha=0.5, label="Sobrecompra (70)")
axes[1].axhline(y=30, color="green", linestyle="--", alpha=0.5, label="Sobreventa (30)")
axes[1].set_title("RSI (14 días)")
axes[1].legend()
axes[1].set_ylim(0, 100)

# SMA Distance
axes[2].bar(data.index, data["SMA_Distance"], color=np.where(data["SMA_Distance"] > 0, "green", "red"), width=1)
axes[2].axhline(y=0, color="black", linewidth=0.5)
axes[2].set_title("Distancia al SMA 200 (% del precio)")

plt.tight_layout()
plt.show()
````

## 5.3 Look-Ahead Bias — El Error #1 en Finanzas con ML

**Look-ahead bias** = usar información del futuro para tomar decisiones del pasado.

Ejemplo: Si el 1 de enero usas datos financieros que la empresa publicó el 15 de marzo,
estás "haciendo trampa" — esos datos no existían el 1 de enero.

**¿Cómo lo evita DeepValueAI?**

1. **Fundamentales**: Usa `publish_date` (fecha real de publicación), no `report_date`
2. **Macro (FRED)**: Usa `realtime_start` (fecha en que la FED publicó el dato)
3. **Sentiment**: Aplica 1 día de lag — noticia del martes solo aparece como feature el miércoles
4. **Split temporal**: Train siempre ANTES de validation, validation ANTES de test
5. **Gaps entre splits**: Espacio de 40 días entre train y val para evitar label leakage

## 5.5 Backtesting — ¿Habría funcionado en la vida real?

El **backtesting** simula cómo habría funcionado tu estrategia si la hubieras usado
en el pasado. Es la prueba más importante antes de invertir dinero real.

**Reglas de DeepValueAI para el backtest:**
- **Entrada**: Modelo dice BUY + precio está cerca de SMA 200
- **Stop Loss**: Si pierde 12% desde la compra → vender todo
- **Trailing Stop**: Si sube 5% y luego cae 8% desde el máximo → vender
- **Take Profit parcial**: Si sube 15% → vender 33% de la posición
- **Tiempo máximo**: Si pasan 120 días sin activar nada → vender
- **Diversificación**: Máximo 10 posiciones abiertas, máximo 15% por ticker
